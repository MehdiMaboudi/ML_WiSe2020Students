```python
import numpy as np
import keras as ks
import keras.backend as kb
from keras import optimizers
from keras.utils import to_categorical
import sklearn
from sklearn.datasets import load_diabetes, load_iris
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
import random
random.seed(57)
sgd_linear = optimizers.SGD(lr=0.00001, momentum=0, nesterov=False)
gdm_linear = optimizers.SGD(lr=0.00001, momentum=0.9, nesterov=False)
nag_linear = optimizers.SGD(lr=0.00001, momentum=0, nesterov=True)
sgd_logistic = optimizers.SGD(lr=0.001, momentum=0, nesterov=False)
gdm_logistic = optimizers.SGD(lr=0.001, momentum=0.9, nesterov=False)
nag_logistic = optimizers.SGD(lr=0.001, momentum=0, nesterov=True)
```

```python
diabetes = sklearn.datasets.load_diabetes(as_frame=True)
diabetes_df = diabetes.frame
X = diabetes.data.to_numpy()
print(X.shape)
y = diabetes.target.to_numpy()

linear = ks.Sequential([
        ks.layers.Dense(10),
        ks.layers.Dense(1)
])


linear.compile(loss='mean_squared_error', optimizer = sgd)
linear_sgd = linear.fit(X,y, epochs = 1000)
linear.compile(loss='mean_squared_error', optimizer = gdm)
linear_gdm = linear.fit(X,y, epochs = 1000)
linear.compile(loss='mean_squared_error', optimizer = nag)
linear_nag = linear.fit(X,y, epochs = 1000)
```

```python
plt.plot(linear_sgd.history['loss'], label='SGD', color = '#2ca0ac')
plt.plot(linear_gdm.history['loss'], label='GDM')
plt.plot(linear_nag.history['loss'], label='NAG')
plt.figsize = (0.5,1000)
plt.title('Linear regression')
plt.ylabel('MSE Value')
plt.xlabel('Number of Iterations')
plt.legend(loc="upper left")
plt.show()
```

```python
plt.plot(linear_gdm.history['loss'], label='GDM')
plt.plot(linear_nag.history['loss'], label='NAG')
plt.figsize = (0.5,1000)
plt.title('Linear regression')
plt.ylabel('MSE Value')
plt.xlabel('Number of Iterations')
plt.legend(loc="upper left")
plt.show()
```

```python
iris = load_iris(as_frame=True)
iris_df = iris.frame
X = iris.data.to_numpy()[50:]
print(X.shape)
y = iris.target.to_numpy()[50:]-1
logistic = ks.Sequential([
        ks.layers.Dense(4),
        ks.layers.Dense(1, activation = 'sigmoid')
])

logistic.compile(loss='binary_crossentropy', optimizer = sgd)
logistic_sgd = logistic.fit(X,y, epochs = 1000)
logistic.compile(loss='binary_crossentropy', optimizer = gdm)
logistic_gdm = logistic.fit(X,y, epochs = 1000)
logistic.compile(loss='binary_crossentropy', optimizer = nag)
logistic_nag = logistic.fit(X,y, epochs = 1000)
```

```python
plt.plot(logistic_sgd.history['loss'], label='SGD', color = '#2ca0ac')
plt.plot(logistic_gdm.history['loss'], label='GDM')
plt.plot(logistic_nag.history['loss'], label='NAG')
plt.title('Logistic regression')
plt.ylabel('Binary Crossentropy Value')
plt.xlabel('Number of Iterations')
plt.legend(loc="upper left")
plt.show()
```
